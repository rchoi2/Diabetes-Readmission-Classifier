# Predicting Diabetes Readmission Using Classification Techniques

The hospital in which you work has a goal to reduce hospital readmission rates for diabetic patients. Both the hospital finance and clinical care teams are interested in how the data science team may help these departments reach this goal.

Develop a model which predicts whether a patient will be readmitted in <30 days. A new diabetic readmission reduction program intervention will use this model in order to target patients at high risk for readmission. Models will be evaluated on AUC, False Positive Rates, True Positive Rates.
You will be expected to present this model to the following teams:

● Hospital health intervention team - how will this affect rates / payment / the bottom line?

● Data science team - how well does this model perform/ how is this model evaluated?

● Data engineering team - what is needed for model implementation?


```python
# Importing Data Analysis and Visualization Packages
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
pd.set_option('max_columns', None)
```

```python
# Read Data 
df = pd.read_csv('./data/diabetic_data.csv')
print(df.shape)
df.head()
```

# Pipeline for Preparing Data 


## Impute Missing Data

### Visualizing Missing Data

```python
df.replace('?',np.nan,inplace = True)
df.isna().sum(axis = 0).sort_values(ascending = False)[:7]
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(df.isnull().T, cbar = False)
```

```python
plt.figure(figsize = (10,8))
missing = pd.DataFrame({'column':df.columns, 'na_percent':df.isnull().sum()/len(df)*100})
missing.sort_values('na_percent', inplace = True)
plt.barh(missing['column'], width = missing['na_percent'])
```

### Addressing Missing Values

```python
# Drop Weight
df.drop(['weight','payer_code'], axis = 1, inplace = True)

# Inpute Medical Specialty with Mode
df.medical_specialty.replace(np.nan, df.medical_specialty.value_counts().index[0], inplace = True)

# Drop Rows with missing values for race, diag_1, diag_2, diag_3
df.dropna(inplace = True)

# Gender 
df = df[df["gender"] != 'Unknown/Invalid']
```

```python
df.isna().sum().sort_values(ascending = False)
```

```python
# Drop ID features / Diag
df.drop(columns = ['encounter_id', 'patient_nbr', 'admission_type_id', 
                   'discharge_disposition_id', 'admission_source_id',
                   'diag_1','diag_2', 'diag_3'], inplace = True)

df['medical_specialty'] = ['InternalMedicine' if x == 'InternalMedicine' 
                           else 'Other' for x in df['medical_specialty']]
```

```python
df.medical_specialty.value_counts()
```

```python
df.info()
```

## Imbalanced Data Set

```python
print(df.select_dtypes('object').columns,'\n')
print(df.select_dtypes('int64').columns)
```

## Recategorize Y 

```python
df['readmitted'] = df['readmitted'].replace(['NO','>30','<30'],['No','No','Yes'])
X = df.drop(columns = ['readmitted'])
y = df['readmitted']

plt.figure(figsize = (10,8))
sns.countplot(y)
```

### Encoding Categorical Variables

```python
X = pd.get_dummies(X, columns = list(X.select_dtypes('object').columns), drop_first = True)
col_names = X.columns
col_names
```

### Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler()
X = X_scaled.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns = col_names)
```

### Oversample

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train
```

## Machine Learning Algorithms / Models

```python
from sklearn.metrics import classification_report, confusion_matrix
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
log_preds = log_model.predict(X_test)
print(classification_report(y_test, log_preds))
print(confusion_matrix(y_test, log_preds))
```

```python
log_model.decision_function(X_train)
```

```python
from sklearn.metrics import roc_curve 

fpr, tpr, thresholds = roc_curve()

def plot_roc_curves(fpr, tpr, label = None):
    plt.plot(fpr, tpr, linewidth = 2, label = label)
    plt.plot([0,1],[0,1],'k--')
    
plot_roc_curves(fpr, tpr)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
y_probas_forest = cross_val_predict(rf_model, X_train, y_train, cv = 3,
                                    method = 'predict_proba')

```

```python
y_scores_rf = y_probas_forest[:,1]
```

```python
rf_preds = rf_model.predict(X_test)
print(classification_report(y_test, rf_preds))
print(confusion_matrix(y_test,rf_preds))
```
